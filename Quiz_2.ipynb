```python
def convert_rgb_to_bw(im_1):
    m,n,k = im_1.shape
    new=np.zeros((m,n),dtype='uint8')
    for i in range (m):
        for j in range(n):
            s=get_0_1_from_triple(im_1[i,j,:])
            new[i,j]=s
    return new
```

```python
def get_value_from_triple(temp):
    #temp=im_1[0,0,:]
    return int(temp[0]/3+temp[1]/3+temp[2]/3)

def convert_rgb_to_gray(im_1):
    m,n,k = im_1.shape
    new=np.zeros((m,n),dtype='uint8')
    for i in range (m):
        for j in range(n):
            s=get_value_from_triple(im_1[i,j,:])
            new[i,j]=s
    return new
```
